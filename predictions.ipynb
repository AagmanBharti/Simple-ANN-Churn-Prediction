```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
model = load_model('model.h5')
```

```python
with open('onehot_encoder_geo.pkl','rb') as file:
    label_encoder_geo = pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    'CreditScore': 700,
    'Geography': 'Spain',
    'Gender': 'Male',
    'Age': 25,
    'Tenure': 3,
    'Balance': 100000,
    'NumOfProducts': 2,
    'HasCrCard': 2,
    'IsActiveMember': 1,
    'EstimatedSalary': 80000
    }
```

```python
geo_encoded = label_encoder_geo.transform(
    [[input_data['Geography']]]
    ).toarray()
```

```python
geo_encoded_df = pd.DataFrame(
    geo_encoded,
    columns=label_encoder_geo.get_feature_names_out(['Geography'])
    )
```

```python
input_df = pd.DataFrame([input_data])
```

```python
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
```

```python
input_df = pd.concat(
    [input_df.drop('Geography', axis=1), geo_encoded_df],axis=1
)
```

```python
input_scaled = scaler.transform(input_df)
```

```python
prediction = model.predict(input_scaled)
```

```python
prediction_proba = prediction[0][0]
```

```python
if prediction_proba > 0.5:
    print(f'The customer is likely to churn with a probability of {prediction_proba:.2f}')
else:
    print(f'The customer is unlikely to churn with a probability of {prediction_proba:.2f}')
```
